```python
import matplotlib.pyplot as plt
import numpy as np

import tensorflow as tf
from tensorflow import keras
import pandas as pd

from keras_preprocessing.image import ImageDataGenerator
```

```python
# data preparation
import random
import glob
import math

ROOT_DIR_PATH = '../data/Data (1)/images/'

LABEL_MAP_PATH = '../data/Data (1)/Classes.xlsx'
TRAIN_CSV_PATH = '../data/Data (1)/train.csv' 
VAL_CSV_PATH = '../data/Data (1)/val.csv'

# read label map xlsx file
df = pd.read_excel(LABEL_MAP_PATH)
class_label_map = pd.DataFrame(df, columns= df.columns)
class_label_map
```

```python
train_df = pd.DataFrame(columns=["image_path", "label"])
val_df = pd.DataFrame(columns= ["image_path", "label"])
test_df = pd.DataFrame(columns= ["image_path", "label"])
```

```python
c = 0
for d in glob.glob(ROOT_DIR_PATH + '/*/'):
    
    #image sampling
    images = glob.glob(d +'/*.jpg')
    dir_name = d.split('/')[-2]
    
#     label = class_label_map[class_label_map["Folder Name "] == dir_name]["Class"].to_numpy()[0]
    label = dir_name
    random.shuffle(images)
    
    print("Total no. of images in folder {}: {}".format(dir_name, len(images)) )
    
    train_num = math.ceil(0.85 * len(images))
    
    train_df = pd.DataFrame(images[: train_num], columns=['image_path'])
    test_df = pd.DataFrame(images[train_num:], columns=['image_path'])
    
    train_df['label'] = [label] * len(train_df)
    test_df['label'] = [label] * len(test_df)
    
    if c == 0:
        header_flag = True
    else:
        header_flag = False
        
    train_df.to_csv('train.csv', mode='a', index= False, header= header_flag)
    test_df.to_csv('test.csv', mode='a', index= False, header= header_flag)
    
    c += 1
     
```

```python
train_df = pd.read_csv('train.csv')
test_df= pd.read_csv('test.csv')

train_df
```

## VGG16

```python
# Hyper params
IMAGE_SIZE = (224,224)
CLASS_MODE = "categorical"
BATCH_SIZE = 5

from tensorflow.keras.applications.vgg16 import preprocess_input
```

```python
gen_tr = ImageDataGenerator(rotation_range= 45,
                            horizontal_flip = True,
                            validation_split= 0.15,
                            preprocessing_function= preprocess_input)

gen_te = ImageDataGenerator(preprocessing_function= preprocess_input)


train_gen = gen_tr.flow_from_dataframe(dataframe= train_df,
                                       directory= None,
                                       x_col= "image_path",
                                       y_col= "label",
                                       subset= "training",
                                       batch_size= BATCH_SIZE,
                                       shuffle= True,
                                       class_mode= CLASS_MODE,
                                       target_size= IMAGE_SIZE
                                      )

valid_gen = gen_tr.flow_from_dataframe(dataframe= train_df,
                                       directory= None,
                                       x_col= "image_path",
                                       y_col= "label",
                                       subset= "validation",
                                       batch_size= BATCH_SIZE,
                                       shuffle= True,
                                       class_mode= CLASS_MODE,
                                       target_size= IMAGE_SIZE
                                      )

test_gen = gen_te.flow_from_dataframe(dataframe= test_df,
                                      directory= None,
                                      x_col= "image_path",
                                      y_col= "label",
                                      batch_size= 1,
                                      shuffle= False,
                                      class_mode= CLASS_MODE,
                                      target_size= IMAGE_SIZE
                                     )
```

```python
# visualizeing train images from generator after preprocess for VGG 16
x,y = next(train_gen)
print(x.shape)

ind = train_gen.class_indices
class_names = list(ind.keys())
 

def plot_images(img, labels):
    plt.figure(figsize=[15, 10])
    for i in range(BATCH_SIZE):
        plt.subplot(5, 5, i+1)
        plt.imshow(img[i])
        plt.title(class_names[np.argmax(labels[i])])
        plt.axis('off')

    return

plot_images(x,y)
```

```python
# visualizeing val images from generator after preprocess for VGG 16
x,y = next(valid_gen)
print(x.shape)

ind = train_gen.class_indices
class_names = list(ind.keys())
 

def plot_images(img, labels):
    plt.figure(figsize=[15, 10])
    for i in range(BATCH_SIZE):
        plt.subplot(5, 5, i+1)
        plt.imshow(img[i])
        plt.title(class_names[np.argmax(labels[i])])
        plt.axis('off')

    return

plot_images(x,y)
```

```python
# VGG16
from tensorflow.keras.applications.vgg16 import VGG16

from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.layers import Dense, Dropout
from livelossplot import PlotLossesKerasTF


base_model = VGG16(weights= 'imagenet', input_shape= (224,224,3))
base_model.trainable = False
#base_model.summary()

output_fc2 = base_model.get_layer("fc2").output


drop_out = Dropout(0.2)(output_fc2)
final_output = Dense(20, activation='softmax')(drop_out)

model = Model(inputs= base_model.input, outputs= final_output)
optimizer = Adam(learning_rate= 0.001)
model.compile(optimizer= optimizer, 
              loss= 'categorical_crossentropy',
              metrics= ['accuracy'])
```

```python
model.summary()
```

```python
n_steps = train_gen.samples // BATCH_SIZE
n_val_steps = valid_gen.samples // BATCH_SIZE
n_epochs = 100

es = EarlyStopping(monitor='val_loss',
                   mode= 'min',
                   patience= 10,
                   restore_best_weights= True)

ckpt = ModelCheckpoint(filepath= 'model_v1.weights.best.hdf5',
                       save_best_only=True,
                       verbose=1)
```

```python
model.fit_generator(train_gen,
                    steps_per_epoch= n_steps,
                    epochs= n_epochs,
                    verbose= 2,
                    callbacks= [ckpt, es, PlotLossesKerasTF()],
                    validation_data= valid_gen)

model.evaluate_generator(valid_gen, n_val_steps)
```
