```python
import shutil
import glob
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import sys
import math
import warnings
import random

from beautifultable import BeautifulTable
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score, accuracy_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.applications import EfficientNetB0, VGG16

from livelossplot import PlotLossesKerasTF 

warnings.filterwarnings('ignore')
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
```

```python
def formatOutputToTable(table_headers, table_data):
    table = BeautifulTable(maxwidth= 100)
    table.column_headers = table_headers
    for i in range(0, len(table_data)):
        table.append_row(table_data[i])

    print(table)
```

**1. (a)Data Exploration and Pre-processing**

```python
DUMP_PATH_TRAIN = '../data/Data (1)/final_images/train/'
DUMP_PATH_TEST = '../data/Data (1)/final_images/test/'
DUMP_PATH_VAL = '../data/Data (1)/final_images/val/'

ROOT_DIR_PATH = '../data/Data (1)/images/'

for d in glob.glob(ROOT_DIR_PATH+ '/*/'):

    images = glob.glob(d + '/*.jpg')
    dir_name = d.split('/')[-2]
    label = dir_name
    random.shuffle(images)

    train_num = math.ceil(0.70 * len(images))
    val_num = math.ceil(0.15 * len(images))
    
    train_img = images[: train_num]
    val_img = images[train_num : train_num+val_num]
    test_img = images[train_num+val_num : ]

    print("Total no. of images in folder {}: {}".format(dir_name, len(images)) )

    os.makedirs(DUMP_PATH_TRAIN + dir_name)
    os.makedirs(DUMP_PATH_VAL + dir_name)
    os.makedirs(DUMP_PATH_TEST + dir_name)

    for img in train_img:
        shutil.copy(img, DUMP_PATH_TRAIN + dir_name)

    for img in val_img:
        shutil.copy(img, DUMP_PATH_VAL + dir_name)

    for img in test_img:
        shutil.copy(img, DUMP_PATH_TEST + dir_name)
```

```python
for files in glob.glob(DUMP_PATH_TRAIN):
    imgs = glob.glob(files + '/*/*')
    print("No of training images: ", len(imgs))


for files in glob.glob(DUMP_PATH_VAL):
    imgs = glob.glob(files + '/*/*')
    print("No of Val images: ", len(imgs)) 


for files in glob.glob(DUMP_PATH_TEST):
    imgs = glob.glob(files + '/*/*')
    print("No of test images: ", len(imgs)) 
```

```python
TRAIN_DIR_PATH = '../data/Data (1)/final_images/train'
VAL_DIR_PATH = '../data/Data (1)/final_images/val'
TEST_DIR_PATH = '../data/Data (1)/final_images/test'

BATCH_SIZE = 5
IMAGE_SIZE = 224 # (224 * 244)
COLOR_MODE = 'rgb'
NUM_CLASSES = 20
```

```python
train_data = tf.keras.utils.image_dataset_from_directory(directory= TRAIN_DIR_PATH,
                                                         shuffle= True,
                                                         batch_size= BATCH_SIZE,
                                                         color_mode= COLOR_MODE,
                                                         image_size= (IMAGE_SIZE, IMAGE_SIZE)
                                                         )

val_data = tf.keras.utils.image_dataset_from_directory(directory= VAL_DIR_PATH,
                                                       shuffle= True,
                                                       batch_size= BATCH_SIZE,
                                                       color_mode= COLOR_MODE,
                                                       image_size= (IMAGE_SIZE, IMAGE_SIZE)
                                                       )

test_data = tf.keras.utils.image_dataset_from_directory(directory= TEST_DIR_PATH,
                                                        shuffle= False,
                                                        batch_size= 1,
                                                        color_mode= COLOR_MODE,
                                                        image_size= (IMAGE_SIZE, IMAGE_SIZE)
                                                        )
```

* Reference Link:
https://keras.io/examples/vision/image_classification_efficientnet_fine_tuning/

```python
resize = tf.keras.Sequential([tf.keras.layers.Resizing(IMAGE_SIZE,IMAGE_SIZE)
                             ])

data_aug = tf.keras.Sequential([tf.keras.layers.RandomFlip("horizontal"),
                                tf.keras.layers.RandomRotation((-0.15,0.2)),
                                tf.keras.layers.RandomContrast(0.2),
                                tf.keras.layers.RandomCrop(200,200),
                                tf.keras.layers.RandomTranslation(0.05,0.05),
                                tf.keras.layers.RandomZoom(0.2,0.2)
                                ])
```

```python
def input_preprocess(image, label):
    label = tf.one_hot(label, NUM_CLASSES)
    return image, label


train_data = train_data.map(input_preprocess,
                            num_parallel_calls= tf.data.AUTOTUNE)

train_data_pr = train_data.prefetch(tf.data.AUTOTUNE)
val_data_pr = val_data.map(input_preprocess)
test_data_pr = test_data.map(input_preprocess)
```

```python
def get_model(name):
    '''
    Given model name, return that specific model
    '''
    input_layer = tf.keras.layers.Input(shape= (IMAGE_SIZE, IMAGE_SIZE, 3))

    
    input_layer = data_aug(input_layer, training= True)
    input_layer = resize(input_layer)

    if name == "VGG16":
        m = VGG16(input_tensor= input_layer)        

    if name == "EFB0":
        m = EfficientNetB0(input_tensor= input_layer)
    
    m.trainable = False
    m = tf.keras.layers.Flatten(name= 'flatten_')(m.layers[-2].output)
    
    m = tf.keras.layers.BatchNormalization()(m)
    m = tf.keras.layers.Dense(1024, activation= 'relu')(m)
    m = tf.keras.layers.Dropout(0.2)(m)
    out = tf.keras.layers.Dense(NUM_CLASSES, activation= 'softmax')(m)

    model = tf.keras.Model(inputs= input_layer, outputs= out)
    optimizer = tf.keras.optimizers.Adam(learning_rate= 1e-5)
    
    model.compile(optimizer= optimizer,
                  loss='categorical_crossentropy',
                  metrics= ['accuracy'])
    print(model.summary())

    return model
```

```python
# get predictions
def get_predicitons(model, data_pr):
    '''
    For a specific model, and a dataset, return all metrics for that dataset.
    '''
    model_preds = model.predict(data_pr)
    model_class_preds = tf.argmax(model_preds, axis= 1)

    true_labels_one_hot = tf.concat([y for x, y in data_pr], axis=0)
    true_label_classes = tf.argmax(true_labels_one_hot, axis= 1)

    cnf_mat = confusion_matrix(true_label_classes, model_class_preds)
    print("MICRO Precision: ", precision_score(true_label_classes, model_class_preds, average= 'micro'))
    print("MICRO Recall: ", recall_score(true_label_classes, model_class_preds, average= 'micro'))
    print("MICRO F1 score: ", f1_score(true_label_classes, model_class_preds, average= 'micro'))
    print()
    print("MACRO Precision: ", precision_score(true_label_classes, model_class_preds, average= 'macro'))
    print("MACRO Recall: ", recall_score(true_label_classes, model_class_preds, average= 'macro'))
    print("MACRO F1 score: ", f1_score(true_label_classes, model_class_preds, average= 'macro'))
    
    print("Accuracy: ", accuracy_score(true_label_classes, model_class_preds))
```

```python
n_epochs = 100
model_name = 'VGG16'
ckpt = tf.keras.callbacks.ModelCheckpoint(filepath= '{}_more_aug.weights.best.hdf5'.format(model_name),
                                          save_best_only=True,
                                          save_weights_only= False,
                                          verbose=1)

es = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                      patience= n_epochs,
                                      restore_best_weights = True)

vgg_model = get_model(model_name)

train_vgg = vgg_model.fit(train_data_pr,
                          epochs= n_epochs,
                          verbose= 1,
                          callbacks= [ckpt, PlotLossesKerasTF()],
                          validation_data= val_data_pr
                          )
```

```python
def my_plot(model):
    plt.plot(model.history['accuracy'])
    plt.plot(model.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()

    plt.plot(model.history['loss'])
    plt.plot(model.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
```

```python
# METRICS on TRAIN DATA

train_data_for_testing = tf.keras.utils.image_dataset_from_directory(directory= TRAIN_DIR_PATH,
                                                         shuffle= False,
                                                         batch_size= BATCH_SIZE,
                                                         color_mode= COLOR_MODE,
                                                         image_size= (IMAGE_SIZE, IMAGE_SIZE)
                                                         )
train_data_pr_test = train_data_for_testing.map(input_preprocess,
                                                num_parallel_calls= tf.data.AUTOTUNE)
get_predicitons(vgg_model, train_data_pr_test)
```

```python
# METRICS on TEST DATA

my_plot(train_vgg)
print(vgg_model.evaluate(test_data_pr))
get_predicitons(vgg_model, test_data_pr)
```

```python
model_name = 'EFB0'
ckpt = tf.keras.callbacks.ModelCheckpoint(filepath= '{}_more_aug.weights.best.hdf5'.format(model_name),
                                          save_best_only=True,
                                          save_weights_only= False,
                                          verbose=1)

es = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                      patience= n_epochs,
                                      restore_best_weights = True)

EFB0 = get_model(model_name)
train_eff = EFB0.fit(train_data_pr,
                     epochs= n_epochs,
                     verbose= 1,
                     callbacks= [ckpt, PlotLossesKerasTF()],
                     validation_data= val_data_pr
                     )
```

```python
# METRICS on TRAIN DATA

get_predicitons(EFB0, train_data_pr_test)
```

```python
# METRICS on TEST DATA

my_plot(train_eff)
print(EFB0.evaluate(test_data_pr))
get_predicitons(EFB0, test_data_pr)
```
