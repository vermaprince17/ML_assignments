**Combined Cycle Power Plant Data Set**

The dataset contains data points collected from a Combined Cycle Power Plant over 6 years (2006-2011), when the power plant was set to work with full load. Features consist of hourly average ambient variables Temperature (T), Ambient Pressure (AP), Relative Humidity (RH) and Exhaust Vacuum (V) to predict the net hourly electrical energy output (EP) of the plant.

Download the Combined Cycle Power Plant data from:
https://archive.ics.uci.edu/ml/datasets/Combined+Cycle+Power+Plant

```python
def formatOutputToTable(table_data, table_headers):
    '''
    prints the input data in table format
    '''
    table = BeautifulTable(maxwidth= 100)
    table.column_headers = table_headers
    for i in range(0, len(table_data)):
        table.append_row(table_data[i])
    print(table)
    
```

**(c) For each predictor, fit a simple linear regression model to predict the response. Describe your results. In which of the models is there a statistically significant association between the predictor and the response? Create some plots to back up your assertions. Are there any outliers that you would like to remove from your data for each of these regression tasks?**

```python

def plotreg(data_X, data_Y, title, feature_name, response_name, slope, intercept, r2Value):
    '''
    Plotting linear regression for single predictor vs response
    '''    
    plt.scatter(data_X, data_Y, s=1)
    
    line_X = np.arange(data_X.min(), data_X.max())
    
    # find predicted responses
    line_Y = slope * line_X + intercept
    
    equation =  feature_name + '*' + str(slope) + '+' + str(intercept)
    
    temp_label = equation + ', R^2=' + str(r2Value)
    plt.plot(line_X, line_Y, 
             label=temp_label, color='black')
    plt.legend(loc='best')
    plt.title(title)
    plt.xlabel(feature_name)
    plt.ylabel(response_name)
    plt.show()
```

```python
def modelLinearRegression(data_X, data_Y):
    '''
    Fitting linear regression model
    '''
    model = sm.OLS(data_Y, data_X).fit()
    coefficients = model.params
    r2_value = round(model.rsquared, 2)
    t_values = model.tvalues
    p_values = model.pvalues
    
    # rounding off t-values, p-values, coeficients, 
    for j in range(0, len(coefficients)):
        coefficients[j] = round(coefficients[j], 3)
        t_values[j] = round(t_values[j], 2)
        p_values[j] = round(p_values[j], 2)
        
    return coefficients, r2_value, t_values, p_values, model
```

```python
feature_names = feature_names_and_responses[:-1]
response = feature_names_and_responses[-1]
univariate_regression_coefficients = {}
print_table = []
for i in range(len(feature_names)):
    feature_name = feature_names[i]
    data_X = np.array(CCPP_data[feature_name])
    data_X = np.reshape(data_X, (len(data_X), 1))
    data_X = sm.add_constant(data_X)
    data_Y = np.array(CCPP_data[response])
    
    # fitting linear model
    coefficients, r2_value, t_values, p_values, model = modelLinearRegression(data_X, data_Y)
    univariate_regression_coefficients[feature_name] = coefficients[1]
    
    # plot reg line for specific predictor
    title = 'Linear Regression for predictor: ' + feature_name
    plt.figure()
    plotreg(data_X[:,1], data_Y, title, feature_name, response, coefficients[1], coefficients[0], r2_value)
    
    infer = ''
    if p_values[1]<=0.05:
        infer = 'good significance'
    else:
        infer = 'poor significance'
    
    # detecting outliers
    
    sea.boxplot(CCPP_data[feature_name])
    plt.show()
    
    #print to table format
    print_table.append([feature_name, coefficients[1], t_values[1], p_values[1], infer])

table_header = ['Feature', 'Coefficient', 't-value', 'p-value', 'Inference']
formatOutputToTable(print_table, table_header)
```

**Conclusion**

The linear regression for various predictors -

1. AT : the p value from the table is 0.0 hence AT has good significance. The coefficient is -2.171, the slope is negative. The box plot shows that there are **no outliers**.

2. V : the p value from the table is 0.0 hence V has good significance. The coefficient is -1.168, the slope is negative. The box plot shows that there are **no outliers**.

3. AP : the p value from the table is 0.0 hence AP has good significance. The coefficient is 1.49 so the slope is positive. The box plot shows that there are **many potential outliers**.

4. RH : the p value from the table is 0.0 hence RH has good significance. The coefficient is 0.456 so the slope is positive. The box plot shows that there are **many potential outliers**.

**(d) Fit a multiple regression model to predict the response using all of the predictors. Describe your results. For which predictors can we reject the null hypothesis H0: βj = 0?**

```python
multivariate_regression_coefficients = {}
print_table = []

data_X = np.array(CCPP_data[feature_names])
data_X = np.reshape(data_X, (len(data_X), len(feature_names)))
data_X = sm.add_constant(data_X)
data_Y = np.array(CCPP_data[response])

# fitting linear model
coefficients, r2_value, t_values, p_values, model = modelLinearRegression(data_X, data_Y)

for i in range(len(feature_names)):
    feature_name = feature_names[i]
    multivariate_regression_coefficients[feature_name] = coefficients[i+1]
    
    infer = ''
    if p_values[i+1] <= 0.05:
        infer = 'Good Significance, reject the null Hypothesis'
    
    else:
        infer = 'Poor Significance, failed to reject null hypothesis'
    
    print_table.append([feature_name, coefficients[i+1], t_values[i+1], p_values[i+1], infer])

table_header = ['Feature', 'Coefficient', 't-value', 'p-value', 'Inference']
formatOutputToTable(print_table, table_header)
print('R**2 value for multivariate linear regression: ',r2_value)
```

**Conclusion**

Clearly from the above table, we can see that all the features are having significance wrt the dependent variable PE as the P value for all the features is less than 0.05.
Thus the Beta values for each of the features is large enough and hence we can reject the null hypothesis for all the features.

```python
univariate_poly_regression_coeff = {}

for i in range(len(feature_names)):
    print_table = []
    
    feature_name = feature_names[i]
    data_X = np.array(CCPP_data[feature_name])
    data_X = np.reshape(data_X, (len(data_X), 1))
    
    t = preprocessing.PolynomialFeatures(3, include_bias= False)
    transformed_X = t.fit_transform(data_X)
    
    transformed_X = sm.add_constant(transformed_X)
    data_Y = np.array(CCPP_data[response])
    
    # fitting linear model
    coefficients, r2_value, t_values, p_values, model = modelLinearRegression(transformed_X, data_Y)
    univariate_poly_regression_coeff[feature_name] = coefficients[1]
    
    temp_name = ''
    for k in range(1, 4):
        feature_name = feature_names[i] + '**' + str(k)
        temp_name += feature_name + ('+' if k <4 else '') 
        
        infer = ''
        if p_values[k]<=0.05:
            infer = 'good significance'
        else:
            infer = 'poor significance'
        
        print_table.append([feature_name, coefficients[k], t_values[k], p_values[k], infer])
    
    # plot reg line for specific predictor
    title = 'Non Linear Regression for predictor: ' + temp_name
    plt.figure()
    sea.regplot(x= feature_names[i], y= response, data= CCPP_data,
                color= 'b',scatter_kws= {"color": colors[i],"s": 5} ,
                order= 3)
    plt.show()

    table_header = ['Feature', 'Coefficient', 't-value', 'p-value', 'Inference']
    print('Univariate Polynomial regression for feature:', feature_names[i])
    formatOutputToTable(print_table, table_header)
    print('R ** 2 Value with', feature_names[i],'predictor:', r2_value)
    print()
```

**(h) Can you improve your model using possible interaction terms or nonlinear associations between the predictors and response? Train the regression model on a randomly selected 70% subset of the data with all predictors. Also, run a regression model involving all possible interaction terms and quadratic nonlinearities, and remove insignificant variables using p-values (be careful about interaction terms). Test both models on the remaining points and report your train and test MSEs.**

```python
# have added random state so that graders can verify the output each time they run this or next following cells.

data_train_X, data_test_X, data_train_Y, data_test_Y = train_test_split(CCPP_data[feature_names],
                                                                        CCPP_data[response],
                                                                        train_size= 0.7,
                                                                        test_size= 0.3,
                                                                        random_state= 42)

all_features = '+'.join(feature_names)
train_NP = data_train_X.join(data_train_Y)
train_DF = pd.DataFrame(train_NP)

simple_linear_model = sml.ols('{0} ~ {1}'.format(response, all_features),
                              data= train_DF)

out_simple_linear_model = simple_linear_model.fit()

y_pred = out_simple_linear_model.predict(data_test_X)
test_error = mean_squared_error(data_test_Y, y_pred)

print("Train Error with all features: ", out_simple_linear_model.mse_resid)
print("Test Error with all features: ", test_error)
```

**Since we have found out in our ealier questions that (AT * AP) and (V * RH) and V ** 2 are insignificant, so we have to drop these terms in the next construction of hypothesis for new model**

```python
combined_features = all_features + '+' + 'I(RH*AP)+I(AT*V)+I(AT*RH)+I(V*AP)+I(AT**2)+I(AP**2)+I(RH**2)'
combined_reg_model = sml.ols('{0} ~ {1}'.format(response, combined_features),
                             data= train_DF)
out_combined_reg_model = combined_reg_model.fit()
print(out_combined_reg_model.summary())
```

**From the above table, it is clear that the P-value of all the features defined in this new model has P-values <= 0.05, thus there are no insignificant predictors in this new model**

```python
combined_features_y_pred = out_combined_reg_model.predict(data_test_X)
test_error_combined_features = mean_squared_error(data_test_Y, combined_features_y_pred)

print("Train Error with only significant features: ", out_combined_reg_model.mse_resid)
print("Test Error with only significant features: ", test_error_combined_features)
```

```python
print_table = []

print_table.append(['Train Error with all features', out_simple_linear_model.mse_resid])
print_table.append(['Test Error with all features', test_error])
print_table.append(['Train Error with only significant features', out_combined_reg_model.mse_resid])
print_table.append(['Test Error with only significant features', test_error_combined_features])

table_header = ['Experiments', 'Results']
formatOutputToTable(print_table, table_header)
```

**(i) KNN Regression using Raw features**

```python
def KNNreg(data_train_X, data_train_Y, data_test_X, data_test_Y, isNorm= False, scale=None):
    '''
    fitting KNN regression model for training dataset
    ret: returns a training_k_dictionary
    '''
    data_train_X = np.array(data_train_X)
    data_train_Y = np.array(data_train_Y)
    data_test_X = np.array(data_test_X)
    data_test_Y = np.array(data_test_Y)
    
    train_k_dict = {}
    for K in range(1, 101):
        KNN = KNeighborsRegressor(n_neighbors= K)
        KNN.fit(data_train_X, data_train_Y)
        y_pred = KNN.predict(data_test_X)
        
                   
        mse = mean_squared_error(data_test_Y, y_pred)
        train_k_dict[1/K] = mse
    
    return train_k_dict
        
```

```python
def plot_train_test_error(train_error_dict, test_error_dict,
                          train_label, test_label):
    '''
    plots the training and test error with corrosponding titles and labels
    '''
    #plot 1/K vs all training errors
    plt.plot(list(train_error_dict.keys()), list(train_error_dict.values()), 'r', label= train_label)
        
    # plot 1/K vs all testing errors
    plt.plot(list(test_error_dict.keys()), list(test_error_dict.values()), 'b', label= test_label)
    
    plt.title('Error vs 1/K value using ')
    plt.ylabel('Mean Square Error')
    plt.xlabel('1/K value')
    plt.legend()
    plt.show()
```

```python
train_error_dict = KNNreg(data_train_X,
                          data_train_Y,
                          data_train_X,
                          data_train_Y)

test_error_dict = KNNreg(data_train_X,
                         data_train_Y,
                         data_test_X,
                         data_test_Y)

plot_train_test_error(train_error_dict,
                      test_error_dict, 
                      'Train error', 
                      'Test error')
```

```python
test_error_dict_sorted = sorted(test_error_dict.items(), key=operator.itemgetter(1))
best_K = int(round(1/test_error_dict_sorted[0][0]))
print("Best K: " + str(best_K) + " with MSE: " + str(test_error_dict_sorted[0][1]))
```

**(i) KNN Regression using Normalized input features**

```python
# Normalize input features
scale = preprocessing.StandardScaler()

scaled_data_train_X = scale.fit_transform(data_train_X)
scaled_data_test_X = scale.fit_transform(data_test_X)
normalized_data_train_X = pd.DataFrame(scaled_data_train_X)
normalized_data_test_X = pd.DataFrame(scaled_data_test_X)


train_error_normalized_dict = KNNreg(normalized_data_train_X,
                          data_train_Y,
                          normalized_data_train_X,
                          data_train_Y)

test_error_normalized_dict = KNNreg(normalized_data_train_X,
                         data_train_Y,
                         normalized_data_test_X,
                         data_test_Y)

plot_train_test_error(train_error_dict,
                      test_error_dict, 
                      'Train error', 
                      'Test error')
```

```python
test_error_normalized_dict_sorted = sorted(test_error_normalized_dict.items(), key=operator.itemgetter(1))
best_K_norm = int(round(1/test_error_normalized_dict_sorted[0][0]))
print("Best K: " + str(best_K_norm) + " with MSE for normalized input features: " 
      + str(test_error_normalized_dict_sorted[0][1]))
```

```python
print_table = []

table_header = ["Model", "Test MSE"]
print_table.append(["Normal Regression Model with all features", test_error])
print_table.append(["Regression Model with only significant features", test_error_combined_features])
print_table.append(["KNN Regression not normalized", test_error_dict_sorted[0][1]])
print_table.append(["KNN Regression with normalizing input features", test_error_normalized_dict_sorted[0][1]])

formatOutputToTable(print_table, table_header)
```

**Conclusion**

The test MSE is almost same for both KNN regression as well as Linear Regression. But the major difference is that in Linear Regression we get an intuiton behind importance of each input feature (significance based on P-value) which is not possible in case of KNN regression.
