**1. Decision Trees as Interpretable Models**

**1. (a)Download the Accute Inflamations data from https://archive.ics.uci.edu/ml/datasets/Acute+Inflammations**

```python
INPUT_FILE_PATH = "../data/Homework 5 Data/diagnosis.data"

df_raw_input = pd.read_csv(INPUT_FILE_PATH, encoding='utf-16',sep='\t', 
                           names=['Temperature','Nausea','Lumbar pain',
                                  'Urine pushing','Micturition pains','Urethra',
                                  'Bladder','Pelvis']
                          )
df = df_raw_input.copy()
temp = df["Temperature"].str.replace(",",".").astype('float')

df["Temperature"] = temp
df.head()
```

```python
X = df.iloc[:, :6]
Y = df.iloc[:, -2:]


print("input: \n", X.head())
print("targets: \n", Y.head())
```

```python
# X_one_hot = pd.get_dummies(X)
# X_one_hot

for col in X.columns[1:]:
    X[col] = X[col].map({'yes': 1, 'no': 0})
for col in Y.columns:
    Y[col] = Y[col].map({'yes': 1, 'no': 0})
    
print(X.head())
print(Y.head())
```

**1. (b) Build a decision tree on the whole data set and plot it.**

```python
# building a decision tree classifier
dec_tree = tree.DecisionTreeClassifier(random_state= 0)
dec_tree.fit(X, Y)

# create a plot to visualize
plt.figure(figsize= (30,20))
plot_tree(dec_tree,
          feature_names= X.columns,
          class_names= Y.columns,
          rounded= True,
          filled= True,
          fontsize= 20
         )
plt.show()
```

**1. (c) Convert the decision rules into a set of IF-THEN rules**

```python
print(dec_tree.classes_)
```

```python
tree_rules = export_text(dec_tree,
                         feature_names = list(X.columns)
                        )
print(tree_rules)
```

```python
def tree_to_pseudo(tree, feature_names):
    '''
    Outputs a decision tree model as if/then pseudocode
    Parameters:
    -----------
    tree: decision tree model
    The decision tree to represent as pseudocode
    feature_names: list
    The feature names of the dataset used for building the decision tree
    '''
    left = tree.tree_.children_left
    right = tree.tree_.children_right
    threshold = tree.tree_.threshold
    features = [feature_names[i] for i in tree.tree_.feature]
    value = tree.tree_.value
    
    def recurse(left, right, threshold, features, node, depth=0):
        indent = " " * depth
        if (threshold[node] != -2):
            print(indent, "if ( " + features[node] + " <= " + str(threshold[node]) + " ) {")
            if left[node] != -1:
                recurse (left, right, threshold, features, left[node], depth+1)
                print(indent,"} else {")
                if right[node] != -1:
                    recurse (left, right, threshold, features, right[node], depth+1)
                    print(indent,"}")
        else:
            print(indent,"return \n" + str(value[node]))
    
    recurse(left, right, threshold, features, 0)
```

```python
tree_to_pseudo(dec_tree, X.columns)
```

**1. (d) Use cost-complexity pruning to find a minimal decision tree and a set of decision rules with high interpretability**

```python
path = dec_tree.cost_complexity_pruning_path(X, Y)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

fig, ax = plt.subplots()
ax.plot(ccp_alphas[:-1], impurities[:-1], marker="o", drawstyle="steps-post")
ax.set_xlabel("effective alpha")
ax.set_ylabel("total impurity of leaves")
ax.set_title("Total Impurity vs effective alpha for training set")
```

```python
mses = []
for ccp_alpha in ccp_alphas:
    
    clf = tree.DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    CV = model_selection.RepeatedKFold(n_splits=5,n_repeats=3)
    score = -1 * model_selection.cross_val_score(clf, X, Y, cv= CV, scoring= "neg_mean_squared_error").mean()
    
    mses.append(score)
    

plt.plot(mses)
plt.xlabel('alphas')
plt.ylabel('MSE')
plt.title('MSE v/s alphas')

best_alpha = ccp_alphas[mses.index(min(mses))]
print("The best alpha value is : {}".format(best_alpha))
```

```python
best_dec_tree = tree.DecisionTreeClassifier(ccp_alpha=best_alpha).fit(X,Y)
tree_to_pseudo(best_dec_tree, X.columns)
```

**2. (i) Fit a PCR model on the training set, with M (the number of principal components) chosen by cross-validation. Report the test error obtained.**

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
error_pcr = {}

for m in range(1, len(features)+1):
    k_fold = KFold(n_splits=10, shuffle=False)
    train_np_X = np.array(train_X)
    train_np_Y = np.array(train_Y)
    
    cv_err = []
    for train_index, val_index in k_fold.split(train_np_X, train_np_Y):
        
        train_temp_X = []
        train_temp_Y = []
        
        for idx in train_index:
            train_temp_X.append(train_X.iloc[idx]) 
            train_temp_Y.append(train_Y.iloc[idx]) 
        
        train_temp_X_df = pd.DataFrame(train_temp_X, columns= features)
        train_temp_Y_df = pd.DataFrame(train_temp_Y, columns= ['ViolentCrimesPerPop'])
        
        val_temp_X = []
        val_temp_Y = []
        
        for idx in val_index:
            val_temp_X.append(train_X.iloc[idx]) 
            val_temp_Y.append(train_Y.iloc[idx]) 
            
        val_temp_X_df = pd.DataFrame(val_temp_X, columns= features)
        val_temp_Y_df = pd.DataFrame(val_temp_Y, columns= ['ViolentCrimesPerPop'])
        
        pca = PCA(n_components= m)
        pca = pca.fit(train_temp_X_df)
        
        new_train_X = pca.transform(train_temp_X_df)
        new_val_X = pca.transform(val_temp_X_df)
        
        pcr_model = LinearRegression()
        pcr_model = pcr_model.fit(new_train_X, train_temp_Y_df)
        
        cv_pred_Y = pcr_model.predict(new_val_X)
        gt_val_Y = np.array(val_temp_Y_df)
        
        err = mean_squared_error(gt_val_Y, cv_pred_Y)  
        cv_err.append(err)
        
    error_pcr[m] = np.mean(cv_err)
```

```python
print('Cross Validation MSE for different M values')
pcr_df_temp = pd.DataFrame([error_pcr])
pcr_df_temp.T
```

```python
plt.plot(list(error_pcr.keys()), list(error_pcr.values()))
plt.xlabel('Principal Components')
plt.xlim(left= -1)
plt.ylabel('Cross-Validation MSE')
plt.title('Principal Component VS MSE')
plt.show()
```

```python
err_sorted = sorted(error_pcr.items(), key= operator.itemgetter(1))
best_M = err_sorted[0][0]
print('BEST M:', best_M)
```

```python
final_pca = PCA(n_components= best_M)
pca = final_pca.fit(train_X)
        
trans_train_X = pca.transform(train_X)
trans_test_X = pca.transform(test_X)

final_pcr_model = LinearRegression()
final_pcr_model = final_pcr_model.fit(trans_train_X, train_Y)
        
pred_Y = final_pcr_model.predict(trans_test_X)        
pcr_mse = mean_squared_error(test_Y, pred_Y)

print('Test MSE of PCR: ', pcr_mse)
```
